# Load the libraries


```
from wikidataintegrator import wdi_core, wdi_login
from getpass import getpass
```

# Login to Wikidata

```
WBUSER = getpass(prompt="username:")  
WBPASS = getpass(prompt='Enter your password: ')  
login = wdi_login.WDLogin(WBUSER, WBPASS)
```

# Get the Wikidata item as json object

```
item = wdi_core.WDItemEngine(wd_item_id='Q42')
json = item.get_wd_json_representation()

```

# Parse
With the item loaded in memory through its json representation it is possible to traverse the item programmatically. For example: List all properties from that item:

```
json["claims"].keys()
```
